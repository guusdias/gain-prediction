# 00 - Setup do Ambiente

Este notebook instala as dependências necessárias do projeto a partir do arquivo `requirements.txt`.

> Observação: em muitos ambientes (VS Code, Jupyter, Google Colab) o comando `!pip` instala os pacotes no ambiente atual do notebook.


```python
import sys
from pathlib import Path

root_dir = Path().resolve()
req_path = root_dir / "requirements.txt"

if not req_path.exists():
    req_path = root_dir.parent / "requirements.txt"

print("Usando arquivo de dependências:", req_path)

!{sys.executable} -m pip install -r "{req_path}"
```
